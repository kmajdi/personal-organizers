```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("cibc.csv", names=["date", "dest", "amt", "amt_recv", "card_no"])
data['date'] = pd.to_datetime(data['date'])
data['dest'] = data["dest"].str[:8]
data['amt'] = data["amt"] - data["amt_recv"].fillna(0)
data.drop("amt_recv", axis=1, inplace=True)
```

```python
freq_dests = data["dest"].value_counts()
freq_dests = freq_dests[freq_dests >= 5]
```

```python
sns.lineplot(data=data[data["dest"].str[:8].isin(freq_dests.index)].groupby(pd.Grouper(key='date', freq='M'))['amt'].sum().reset_index().sort_values('date'), x="date", y="amt")
```

```python
sns.lineplot(data=data[data["dest"].str[:8].isin(freq_dests.index)].groupby(['dest', pd.Grouper(key='date', freq='M')])['amt'].sum().reset_index().sort_values('date'), x="date", y="amt", hue="dest")
```
